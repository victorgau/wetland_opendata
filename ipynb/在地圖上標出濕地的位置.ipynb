```
url = "https://github.com/victorgau/wetland_opendata/raw/master/wetlands.csv"
```

```
import folium
```

```
import geocoder
```

```
gps = geocoder.osm("台灣").latlng
```

```
map = folium.Map(location=gps, zoom_start=7)
```

```
map
```

```
df.head()
```

```
df.info()
```

```
for i, row in df.iterrows():
  folium.Marker([row['lat'], row['lng']], popup=f"<div style='white-space:nowrap'>{row['name']}</div>").add_to(map)
```
